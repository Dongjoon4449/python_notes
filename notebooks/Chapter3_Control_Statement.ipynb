## **for문**
- for문은 Java에서 작동하는 방식과 매우 다르다.
- 파이썬에서 for문의 핵심은 기본적으로 리스트를 순회하는 방식이라는 것이다

**for문의 기본 구조:**

for 변수 in 리스트(또는 튜플, 문자열):
    수행할_문장1
    수행할_문장2
    ...

```python
test_list = ['one', 'two', 'three'] 
for i in test_list: 
     print(i) 
```

**for문에서 함께 자주 사용되는 range 함수**
- range 함수는 숫자 리스트를 자동으로 만들어주는 함수다

```python
add = 0 
for i in range(1, 11): 
    add = add + i 
print(add)
```

## **(중요!) for문 list comprehension**

list conprehension의 기본 뼈대:
[표현식 for 변수 in iterable]

- 이걸 말로 풀면: iterable(list 등)에서 표현식을 적용한 결과를 리스트로 만든다

list comprehension에 조건문이 붙은 구조:
[표현식 for 항목 in iterable(반복가능개체) if 조건문]

```python
# 가장 기본적인 사용방식
nums = [1, 2, 3, 4]
result = [x * 2 for x in nums]
print(result)

# 조건이 있는 경우 (필터링)
evens = [x + 3 for x in nums if x % 2 == 0]
print(evens)

# 문자열 처리 예제
names = ["Alice ", "  Bob", "Kim   "]
clean = [n.strip().lower() for n in names]
print(clean)
```

**zip 함수로 여러 리스트 함께 순회하기**
- 두 개 이상의 리스트를 동시에 순회하고 싶을 때는 zip 함수를 사용한다.

```python
names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name}: {score}점")
```
